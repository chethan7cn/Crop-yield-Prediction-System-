```python
#Reading the clean data file into data frame :

crop_df = pd.read_csv("C:\Ibridge_project\Practice_File\Clean_Data\crop_data_df6.csv")

crop_df
```

```python
#Converting the data frame into dic :

crop_df_dic = crop_df.to_dict(orient='records')
crop_df_dic
```

```python
#Define the SQL query :
insert_query = text("""  
        INSERT INTO Crops(CROP_ID, CROP_NAME, CROP_TYPE, GROWING_SEASON)
        VALUES ( :CROP_ID, :CROP_NAME, :CROP_TYPE, :GROWING_SEASON)
        """
)

#Inserting data into data base :
with engine.connect() as connection :
    connection.execute(insert_query,crop_df_dic)
    connection.commit()
    print("Data inserted successfully")
```
